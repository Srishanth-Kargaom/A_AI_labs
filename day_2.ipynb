**Identity Operators**

"is" & "is not"

```python
l1 = [1,2,3,4,5]
l2 = [1,2,3,4,5]

print(id(l1))
print(id(l2))

l3 = l1
print(id(l3))

print(l2 is l1)
print(l3 is l1)
print(l1 is not l2)
```

**Unicode Representation**

```python
alphaNumeric_string = "abcdefghxyzABCDEFGHXYZ0123456789"

for c in alphaNumeric_string:
  print("character: ", c)
  print("unicode value: ", ord(c))
```

```python
name = "srishanth"

for c in name:
  print("character: ",c)
  print("unicode value: ", ord(c))
  print("hex value: ",hex(ord(c)))
```
