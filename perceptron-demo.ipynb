```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('accidents.csv', skiprows=1)
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(
    data=df,
    x='car_safety_rating',
    y='driver_sleeping_hours',
    hue='accident_happened'
)
```

```python
X = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X, y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
