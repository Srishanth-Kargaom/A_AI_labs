```python
# defining a function

def python_function():
  print("Welcome to Functions class")
  l = [100,200,300,400]
  return l

# calling a function

func_return = python_function()
print(func_return)
print(type(func_return))
```

```python
# passing arguments

def f_name(name):
  print("Entered Name : {}".format(name))
  return name*2

ret_val = f_name("Virat")
print(ret_val)
print(type(ret_val))
```

```python
ret_val = f_name(10)
print(ret_val)
print(type(ret_val))
```

```python
ret_val = f_name([1,2,3])
print(ret_val)
print(type(ret_val))
```

```python
# default arguments
def statement(msg, times=1):
  print(msg*times)

statement("Hello")
statement("World", 5)
```

```python
# Global variables vs Local Variables

def func_a():
  a = 100
  print("Local Variable - a: ", a)

func_a()
print("Access Local Variable - a from outside: ", a)
```

```python
# global variable
b = 100

def func_b():
  print("Access Global Variable -b Inside Function: ", b)

func_b()
print("Access Global Variable - b Outside Function: ", b)
```

```python
# purpose of "global" keyword

def func_d():
  global d
  print("Global Variable d : ", d)
  d += 500
  print("Changing d inside the function: ",d)

d = 1000
print("Initializing d: -->", d)
func_d()
print("Outside function, d changed to : ", d)
```

```python
# nested functions

def outer():
  x = 10
  def inner():
    print("x inside inner:", x)
  inner()

outer()
```

```python
# nonlocal keyword

def outer():
  msg = "outer"
  def inner():
    nonlocal msg
    msg = "modified by inner"
  inner()
  print(msg)

outer()
```

```python
class Stu:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def avg(self):
        return sum(self.marks) / len(self.marks)

s = Stu("srishanth", [95, 99, 85])

print(s.name)
print("Average marks:", s.avg())
```

```python
# Defining a base class for ATM
class ATM:
    def __init__(self, account_number, pin):
        self.account_number = account_number
        self.pin = pin
        self.balance = 0

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"Withdrew â‚¹ {amount}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited â‚¹ {amount}")

    def transfer(self, amount, recipient_account):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            recipient_account.balance += amount
            print(f"Transferred â‚¹ {amount} to acc no : {recipient_account.account_number}")


# Defining a concrete class for a UserAccount
class UserAccount(ATM):
    def __init__(self, account_number, pin, balance):
        super().__init__(account_number, pin)
        self.balance = balance


# Defining a concrete class for a BusinessAccount
class BusinessAccount(ATM):
    def __init__(self, account_number, pin, balance):
        super().__init__(account_number, pin)
        self.balance = balance

    def transfer(self, amount, recipient_account):
        # Override the transfer method to include a commission fee
        super().transfer(amount, recipient_account)
        print(f"Commission fee applied to transfer of â‚¹{amount}")


# Creating an instance of UserAccount
user_account = UserAccount("1234567890", "1234", 1000.0)

# Creating an instance of BusinessAccount
business_account = BusinessAccount("4092430", "1234", 5000.0)

# Perform transactions
user_account.withdraw(500.0)
business_account.deposit(1000.0)
user_account.transfer(200.0, business_account)
```
