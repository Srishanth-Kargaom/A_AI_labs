# **Prompts in LangGraph**
Nodes can contain full LangChain chains — you're not limited to a single `llm.invoke()` call.
Here we attach a `ChatPromptTemplate` inside a node before calling the LLM.

```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
from dotenv import load_dotenv
import os

load_dotenv()

if os.environ.get('GROQ_API_KEY'):
    print("Groq API Key is set.")
else:
    raise ValueError("GROQ_API_KEY not found.")
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)
```

### **Create a Prompt Template**
`ChatPromptTemplate.from_messages()` lets you define a multi-turn prompt with placeholders.
The {topic} placeholder is filled in at invoke time.

```python
from langchain_core.prompts import ChatPromptTemplate

my_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that helps to create and curate posts for a social media platform."),
    ("human",  "I want to create a post about {topic}")   # {topic} is the dynamic variable
])
```

### **Chain With Prompts**
The `|` operator wires the prompt template and the LLM into a single `Runnable`.
`chain.invoke({...})` fills the template then calls the LLM.

```python
# prompt_template → llm (the LCEL pipe syntax)
chain = my_prompt | llm

# invoke() fills {topic} and sends the rendered messages to the LLM
response = chain.invoke({"topic": "The importance of data privacy in the digital age"})
print(response.content)
```
