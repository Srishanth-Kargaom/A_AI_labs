```python
# langchain_groq gives us the ChatGroq class
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv  # reads .env file and puts vars into os.environ

# load_dotenv() scans for a .env file and adds its key=value pairs to env variables
# so GROQ_API_KEY=xxx in .env becomes accessible as os.environ['GROQ_API_KEY']
load_dotenv()

# Quick sanity-check: confirming the key is loaded before making any API calls
if os.environ.get("GROQ_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("GROQ_API_KEY not found — add it to your .env file")
```

```python
# ChatGroq is a LangChain chat model that calls the Groq inference API
# model: which LLM to use on Groq (llama-3.3-70b-versatile is fast & capable)
# temperature=0: deterministic output — same input → same output every time
my_llm_groq = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

# .invoke() sends the string as a HumanMessage and returns an AIMessage object
# .content extracts just the string reply from that AIMessage
my_llm_groq.invoke("tell me a fun fact about Aditi consulting").content
```

```python
# Alternative: init_chat_model is a provider-agnostic factory
# It auto-detects the provider from the model string or an explicit provider= arg
# Useful when you want to swap providers without changing downstream code
from langchain.chat_models import init_chat_model

# provider="groq" tells LangChain which backend to use
llm_groq2 = init_chat_model(model="llama-3.3-70b-versatile", model_provider="groq")

# Returns a full AIMessage object (not just .content) so you can inspect metadata too
llm_groq2.invoke("Hello, how are you?")
```
