### **Lists & Tuples**

#### **Lists**

1. Initializing a list
2. Accessing elements & Slicing
3. Adding elements
   * append
   * insert
   * extend
4. Updating list items
5. Remove, Del, Clear, Pop
6. List Comprehensions

```python
l = []
print(l)
```

```python
l = list()
print(l)
```

```python
name = "Virat Kohli"
l = list(name)
print(l)
```

```python
l = [1,2,3,"a","b","c","rcb","gate","da"]

print(l[4])
print(l[-4])
```

```python
l = [[1,2,3,[4,5]],[23,[12,34],9],30,21]

print(len(l))
```

```python
print(l[0][3][0])
print(l[0][3][-2])
```

```python
# slicing a list

l = ["sachin","sehwag","gauti","virat","yuvi","raina","dhoni","zaheer","bumrah"]

print(len(l))
```

```python
l[2:5]
```

```python
l[::-1]
```

```python
l[-4:2]
```

```python
alpha = ['a','b','c','d','e','f','g','h','i','j']
alpha[::2]
```

```python
# append
l = [1,2,3]
l.append(4)
print(l)
```

```python
# insert
l.insert(1,100)
print(l)
```

```python
# extend
l.extend([10,20,30])
print(l)
```

```python
# remove, pop, del
l = [1,2,3,4,5]
l.remove(3)
print(l)
```

```python
popped = l.pop()
print(popped)
print(l)
```

```python
del l[0]
print(l)
```

```python
# List Comprehension
squares = [x**2 for x in range(1,11)]
print(squares)
```

```python
evens = [x for x in range(1,21) if x%2==0]
print(evens)
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [val for row in matrix for val in row]
print(flat)
```
