### **Filter**

```python
scores = [50,45,10,76,100,12,5]

def func(x):
  if x>40:
    return True
  else:
    return False

filtered_list = filter(func, scores)
print(filtered_list)
```

```python
filtered_list = list(filter(func, scores))
print(filtered_list)
```

```python
numbers = [12,42,3,75,124,29,71,1,2,47,10001]
evenNumbers = list(filter(lambda x: x%2==0, numbers))
oddNumbers = list(filter(lambda x: x%2==1, numbers))
print(evenNumbers)
print(oddNumbers)
```

```python
def isPrime(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n%i == 0:
      return False
  return True

numbers = [7,17,29,144,142,2001,199,176,187]
primeNumbers = list(filter(isPrime, numbers))
print(primeNumbers)
```

### **MAP Function**

```python
def myFun(a):
  if a>18:
    return False
numbers = [1,2,9,12,18,20,25,40,50]
map_numbers = list(map(myFun, numbers))
print(map_numbers)
```

```python
scores = [55,75,45,38,58,67,81,92]
map_scores = list(map(lambda x: x+7, scores))
print(map_scores)
```

```python
scores = [55,75,45,38,58,67,81,92]
# add 7 if score is >70 else leave as it is
map_scores = list(map(lambda x: x+7 if x>70 else x, scores))
print(map_scores)
```

### **Recursion**

```python
def func_new(a):
  print("Value of a : -->",a)
  if a<=0:
    return a
  else:
    return func_new(a-1)

ret_val = func_new(5)
print("return val: -->", ret_val)
```

```python
def factorial(n):
  print("value of n: -->",n)
  if n==0:
    return 1
  return n*factorial(n-1)

ret_val = factorial(5)
print("ret val: -->", ret_val)
```

```python
def sum_n_numbers(n):
  if n==0:
    return 0
  print("n val : -->",n)
  return n+sum_n_numbers(n-1)

ret_val = sum_n_numbers(5)
print(ret_val)
```

```python
def head_recursion(n):
  if n>0:
    head_recursion(n-1)
    print("n val: -->", n)

def tail_recursion(n):
  if n>0:
    print("n val: -->", n)
    tail_recursion(n-1)

head_recursion(5)
print("---")
tail_recursion(5)
```

```python
# you can have more than 1 function calls in a recursive function

# 1, 1, 2, 3, 5 , 8, 13, 21
# 1  2  3  4   5  6  7   8

# f(n) = f(n-1)+f(n-2)

def getFibonacciNum(n):
  print("Function call n: -->", n)
  if n==1 or n==2:
    return 1
  return getFibonacciNum(n-1)+getFibonacciNum(n-2)
```

```python
getFibonacciNum(5)
```

```python
# GATE DA

def count(child_dict, i):
 if i not in child_dict.keys():
  return 1
 ans = 1
 for j in child_dict[i]:
  ans += count(child_dict, j)
 return ans
child_dict = dict()
child_dict[0] = [1,2]
child_dict[1] = [3,4,5]
child_dict[2] = [6,7,8]
print(count(child_dict,0))
```

```python
def fun(D, s1, s2):
 if s1 < s2:
  D[s1], D[s2] = D[s2], D[s1]
  fun(D, s1+1, s2-1)

d = [100,200,5,6,1000,1,10000,12,500]
print(d)
fun(d, 2,5)
print(d)
```
