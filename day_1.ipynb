```python
# set is unordered, so indexing is not applicable
set_3 = {1,4,2,2.0,2.5,3.7,3,"Virat","Rajini"}
print(set_3)
```

**Frozen Sets**

```python
f_set_1 = frozenset((1,2,1.0,"KKR","CSK","MI"))
print(f_set_1)
print(type(f_set_1))
```

**Dictionary**

```python
rcb_batter_scores = {"Virat":55,"Faf":25,"Green":20,"Maxi":0}
print(rcb_batter_scores)
print(type(rcb_batter_scores))
```

```python
rcb_batter_scores.update({"DK":35})
print(rcb_batter_scores)
```

```python
# duplicate keys - last value wins
aus_batter_scores = {"Warner":75,"Head":100,"Maxi":120,"Maxi":70}
print(aus_batter_scores)
```
